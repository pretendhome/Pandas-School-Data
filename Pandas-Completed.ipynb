```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Desktop/Resources/schools_complete.csv"
student_data_to_load = "Desktop/Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])


school_data_complete.head()
```

```python
per_school_budget = school_data_complete.groupby(['school_name']).mean()["budget"]
```

```python
per_school_counts = school_data_complete.groupby(['school_name']).count()["student_name"]
```

```python
per_school_capita =  per_school_budget / per_school_counts
```

```python
school_type = school_data_complete.groupby('school_name')['type'].unique()
school_type = school_type.str[0]
```

```python
per_school_math = school_data_complete.groupby(['school_name']).mean()["math_score"]
```

```python
per_school_reading = school_data_complete.groupby(['school_name']).mean()["reading_score"]
```

```python
perc_math = school_data_complete[school_data_complete["math_score"]>70]#school_data_complete[[school_data_complete['reading_score']>70].groupby(school_data_complete['school_name']).mean()
per_school_passing_math = perc_math.groupby(['school_name']).mean()["math_score"]
```

```python
perc_reading = school_data_complete[school_data_complete["reading_score"]>70]
per_school_passing_reading = perc_reading.groupby(['school_name']).mean()["reading_score"]
per_school_passing_reading.head
```

```python
overall_passing_rate = (per_school_passing_reading + per_school_passing_math) / 2
```

```python
school_types = school_data_complete.set_index('school_name', inplace=True)
school_summary = pd.DataFrame({"School Name": school_types,
                                    "School Type": school_type,
                                    "Total Students": per_school_counts,
                                    "Total School Budget": per_school_budget,
                                    "Per Student Budget": per_school_capita,
                                    "Average Math Score": per_school_math,
                                    "Average Reading Score": per_school_reading,
                                    "% Passing Math": per_school_passing_math,
                                    "% Passing Reading": per_school_passing_reading,
                                    "% Overall Passing Rate": overall_passing_rate}) 
school_summary
```
